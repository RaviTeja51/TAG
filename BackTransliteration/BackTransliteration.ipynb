# Back Transliteration of Roman into Sanskrit

```python
import re
#Algorithm reference:
#                      TY  - BOOK
#                      AU  - Nair, Jayashree
#                      AU  - Sadasivan, Anand
#                      PY  - 2019/03/30
#                      SP  - 
#                      T1  - A Roman to Devanagari Back-Transliteration Algorithm based on Harvard-Kyoto Convention
#                      ER  - 
#


#unicode source: https://unicode.org/charts/PDF/U0900.pdf


#hash-map for vowels

vowels = {"a":"\u0905","A":"\u0906","i":"\u0907","I":"\u0908",
         "u":"\u0909","U":"\u090A","R":"\u090B","RR":"\u0960",
         "lR":"\u090C","lRR":"\u0961","e":"\u090F","ai":"\u0910",
         "o":"\u0913","au":"\u0914","M":"\u0902","H":"\u0903"}

#hash-map for consonants
consonants ={"k":"\u0915","kh":"\u0916","g":"\u0917","gh":"\u0918","G":"\u0919",
            "c":"\u091A","ch":"\u091B","j":"\u091C","jh":"\u091D","J":"\u091E",
            "T":"\u091F","Th":"\u0920","D":"\u0921","Dh":"\u0922","N":"\u0923",
             "t":"\u0924","th":"\u0925","d":"\u0926","dh":"\u0927","n":"\u0929",
             "p":"\u092A","ph":"\u092B","b":"\u092C","bh":"\u092D","m":"\u092E",
            "y":"\u092F","r":"\u0930","l":"\u0931","v":"\u0935","z":"\u0936",
             "S":"\u0937","s":"\u0938","h":"\u0939"}
#special-characters
sp_char = {"U":"\u0942","u":"\u0941","`":"\u094D","’":"\u093D",}

#vowel_signs
vowel_signs = {"A":"\u093E","i":"\u093F","I":"\u0940","u":"\u0941",
               "U":"\u0942","R":"\u0943","RR":"\u0944","e":"\u0947",
               "ai":"\u0948","o":"\u094B","au":"\u094C"}
```

```python
#storing the text into a list

txt = []
with open("kasyapa2.txt") as f:
    for line in f:
        txt.append(line)



with open("back_transliterate.txt","w",encoding = "utf-8") as f:
    print("Back Transliteration started ...")
    
    for line in txt:
        
        new_line = []
        i = 0
        line_length = len(line)
        prev_char_consonant =  False #previous character is consonant
        
        while i < line_length:
            
            char = line[i]
            
            if char == "C":
                #There is no "C" in Harvard-Kyoto
                char = "S"
            
            
            
           
            
            if char in vowels:
                    
                    
                    if i+1 < line_length:
                        #checking for the next character in the string
                        
                        if char == "a" and line[i+1] in ['i','u']:
                            char += line[i + 1]
                            i += 1
                            
                        if char == "o" and line[i+1] == 'M':
                            char += line[i+1]
                            i += 1
                    
                    if prev_char_consonant:
                        
                        
                        new_line[-1]= new_line[-1].strip(sp_char["`"])
                        if char in ["A","i","I","u","U","R","RR",
                                   "e","ai","o","au"]:
                            '''
                               To cover cases like: व + आ = वा
                               If this 'if block' is not there then
                                व + आ = वआ
                            '''
                            new_line.append(vowel_signs[char])
                        
                    else:
                        new_line.append(vowels[char])
                    
                    prev_char_consonant = False
                    
            else:
                    if char in consonants:
                        if char in ['k','g','j','T','D','t','d','p','b']:
                            
                            if i + 1 < line_length and line[i+1] == 'h':
                                char += 'h'
                        new_line.append(consonants[char] + sp_char["`"])
                        
                        prev_char_consonant = True
                    else:
                        if char in sp_char:
                            new_line.append(sp_char[char])
                        else:
                            new_line.append(char)
            i += 1
        
                        
        f.write("".join(new_line))   
    print("....")
    print("....")
    print("Back Transliteration completed")
                        
                
                
                
    
```
