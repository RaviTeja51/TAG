```python
import pandas as pd
import re
lines = []
with open("kasyapa.txt") as file:
    for i in file:
        lines.append(i)
```

```python
processed_lines = []

pattern = re.compile("\|+") #regular exp for "|"
numbers = re.compile("[0-9]+") #regular exp for "numbers"
spaces = re.compile("(\t)+") #regular exp for removing tabs
new_line = re.compile("\n") #regular exp for removing new lines
hypen = re.compile("\-+") #regualr exp for hypens

for i in lines:
    if i != "\n":
        if pattern.search(i):
            
            line = spaces.sub("",i) #removing extra spaces
            line = hypen.sub("",line) #removing hypens
            line = numbers.sub("",line) # remove numbers
            line = new_line.sub("",line) # removing new line
            line = pattern.split(line) #splitting at |
            
            while "" in line:
                line.remove("") #removing empty character
            processed_lines.extend(line)
            
            
            
            
            
        
        
       
```

```python
for i in processed_lines:
    print(i)
```
